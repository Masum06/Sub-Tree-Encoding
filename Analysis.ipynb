```python
import json
import copy
def getJsonData(JsonFile):
    with open(JsonFile, encoding="utf8") as f:
        data = json.load(f)
    return data

def dump_dict(data, name):
    with open(name, 'w', encoding="utf8") as f:
        json.dump(data, f)
```
